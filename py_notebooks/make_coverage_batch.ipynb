```python
import pandas as pd
```

```python
mutation_frequencies_s1 = pd.read_csv('/Users/lincoln.harris/Desktop/mutation_frequencies_by_gene_s1.csv')
mutation_frequencies_s1
```

```python
mutation_frequencies_s1 = mutation_frequencies_s1[['gene', 'mutation', 'count', 'chrom', 'start', 'end']]
mutation_frequencies_s1
```

```python
mutation_frequencies_s1 = mutation_frequencies_s1.dropna()
mutation_frequencies_s1
```

```python
covDF = pd.DataFrame(columns=['chrom', 'start_pos', 'end_pos', 'outfile'])
covDF
```

```python
chrom_list = []
start_pos_list = []
end_pos_list = []
outfile_list = []

for i in range(0, len(mutation_frequencies_s1.index)):
    row = mutation_frequencies_s1.iloc[i]
    gene = row['gene']
    mut = row['mutation']
    chrom = row['chrom']
    start = row['start']
    end = row['end']
    
    outStr = gene + '_' + mut + '_coverageByCell.csv'
    
    chrom_list.append(chrom)
    start_pos_list.append(start)
    end_pos_list.append(end)
    outfile_list.append(outStr)
```

```python
covDF['chrom'] = chrom_list
covDF['start_pos'] = start_pos_list
covDF['end_pos'] = end_pos_list
covDF['outfile'] = outfile_list
covDF
```
