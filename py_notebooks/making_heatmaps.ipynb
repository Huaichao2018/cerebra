```python
import pandas as pd
pd.set_option('display.max_rows', 300)
pd.options.display.max_columns = None

import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
coverage_x_sample_ROI = pd.read_csv('/Users/lincoln.harris/Desktop/coverage_x_sample_ROI_cerebra.csv')
coverage_x_sample_ROI = coverage_x_sample_ROI.set_index('Unnamed: 0')
coverage_x_sample_ROI
```

```python
v_matrix_ROI = coverage_x_sample_ROI.applymap(lambda x: int(x.split(':')[0]))
t_matrix_ROI = coverage_x_sample_ROI.applymap(lambda x: int(x.split(':')[1]))
```

```python
ratios_matrix_ROI = v_matrix_ROI / t_matrix_ROI
ratios_matrix_ROI = ratios_matrix_ROI.fillna(0)
ratios_matrix_ROI
```

```python
keep_ROIs = ['BRAF_V600E', 'EGFR_L858R', 'EGFR_K745_A750>T', 'KRAS_G12A', 'KRAS_G12C', 'KRAS_G12S', 'KRAS_G12V',
             'KRAS_G13C', 'KRAS_G12F', 'KRAS_G13D', 'KRAS_Q61H', 'BRAF_W450L']
keep_idx = ratios_matrix_ROI.index.isin(keep_ROIs)
ratios_matrix_ROI_trimmed = ratios_matrix_ROI[keep_idx]
ratios_matrix_ROI_trimmed
```

```python
# this seems to work -- doesnt truncate labels
sns.set(font_scale=0.7)
ax = sns.heatmap(ratios_matrix_ROI_trimmed, annot=False, xticklabels=True, yticklabels=True)
ax.figure.subplots_adjust(left = 0.2) # change 0.3 to suit your needs.
plt.savefig('/Users/lincoln.harris/Desktop/fig2d_cerebra.pdf')
```
